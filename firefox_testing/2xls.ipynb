```python
import yaml
import openpyxl as xl
```

```python
data = None
with open("testcases.yml",'r',encoding='utf-8') as f:
    data = list(yaml.load_all(f, Loader=yaml.SafeLoader))[0]['testcases']
```

```python
data
```

```python
wb = xl.Workbook()
ws = wb.active
for n,i in enumerate(data):
    ws['A{}'.format(n+1)] = i['id']
    ws['B{}'.format(n+1)] = i['title']
    s1 = ''
    for nn, s in enumerate(i['prequisition']):
        s1 += "{}) {}；".format(nn,s)
    ws['C{}'.format(n+1)] = s1
    s1 = ''
    for nn, s in enumerate(i['steps']):
        s1 += "{}) {}；".format(nn,s)
    ws['D{}'.format(n+1)] = s1
    s1 = ''
    for nn, s in enumerate(i['expected']):
        s1 += "{}) {}；".format(nn,s)
    ws['E{}'.format(n+1)] = s1
    s1 = '扫尾工作：'
    for nn, s in enumerate(i['cleanup']):
        s1 += "{}) {}；".format(nn,s)
    ws['F{}'.format(n+1)] = s1
wb.save("sample.xlsx")
```
